```python
def plot_read_data(x,y,file):
    import pandas as pd
    with open(file, 'r', encoding='utf-8') as final:
        data = final.read()
    with open(file, 'w', encoding='utf-8') as final:    
        final.write(x+" "+y+'\n')
        final.write(data)
    return data
```

```python
def plots(x,y,file):
    import matplotlib.pyplot as plt
    import array
    import pylab
    data=plot_read_data(x[0],y[0],file[0])
    df=pd.read_csv (file[0], sep=" ")
    plt.plot(df[x[0]],df[y[0]],label=y[0])
    with open(file[0], 'w', encoding='utf-8') as final:
        final.write(data)
    for i in range(1,5):
        data=plot_read_data(x[i],y[i],file[i])
        df=pd.concat([df,pd.read_csv (file[i], sep=" ")],axis=1)
        with open(file[i], 'w', encoding='utf-8') as final:
            final.write(data)
        plt.plot(df[x[i]],df[y[i]],label=y[i])
    plt.legend(fontsize=14)
    return plt
```

```python
file="конт1.txt","конц2.txt","конц3.txt","конц4.txt","конц5.txt"
x="конт1","конц2","конц3","конц4","конц5"
y="конт1_y","конц2_y","конц3_y","конц4_y","конц5_y"
plots(x,y,file)
```
